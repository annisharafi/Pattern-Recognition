```
import pandas as pd
import numpy as np
import nltk
import math
import re
from bs4 import BeautifulSoup
from nltk.tokenize import WordPunctTokenizer
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```
#import data csv

df=pd.read_csv('pengpol.csv',encoding='ISO-8859-1')
df.drop(['No'],axis=1)
```

```
df=df.fillna(' ')
```

```
#initialization token and stopword
token =  WordPunctTokenizer()
stopWord = set(stopwords.words('indonesian')) 
```

```
#preprocessing function
def preprocessing(row):
  clean = row.replace(u'\xa0', u' ')
  clean = re.sub("[^a-zA-Z0-9]", " ", row)
  clean = clean.lower()
  token_word = token.tokenize(clean)
  word=[]
  for i in token_word:
    if (i not in stopWord):
      word.append(i)
  return(" ".join(word).strip())
```

```
#cleaning Gejala
dataGejala=[]
for i in range(len(df)):
  dataGejala.append(preprocessing(df['Gejala'][i]))
```

```
#cleaning Diagnosis
dataDiagnosis=[]
for i in range(len(df)):
  dataDiagnosis.append(preprocessing(df['Diagnosis'][i]))
```

```
#dot product and magnitude vector
def dot_product(vector_x, vector_y):
    dot = 0.0
    for e_x, e_y in zip(vector_x, vector_y):
        dot += e_x * e_y
    return dot

  
def magnitude(vector):
    mag = 0.0
    for index in vector:
        mag += math.pow(index, 2)
    return math.sqrt(mag)
```
